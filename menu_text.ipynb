```python
def yelp_info(price, categories, location):
    price = price
    categories = categories
    SEARCH_LIMIT = 50
    results = []
    parsed_results = []
    
    total_num = get_total(location)

    try:
        for i in range(0, total_num, SEARCH_LIMIT):
            time.sleep(0.7)
            url = 'https://api.yelp.com/v3/businesses/search'

            headers = {
                'Authorization': 'Bearer {}'.format(api_key),
            }

            url_params = {
                'limit': SEARCH_LIMIT,
                'offset': i,
                'price': price.replace(' ', '+'),
                'categories': categories.replace(' ','+'),
                'location': location.replace(' ','+')

            }
            response = requests.get(url, headers=headers, params=url_params)
            print(i)
            results.extend(response.json()['businesses'])

        for business in results:
            parsed_dict = {}
            parsed_dict['name'] =business['name']
            parsed_dict['review_count'] = business['review_count']
            parsed_dict['rating'] = business['rating']
            parsed_dict['price'] = business['price']
            parsed_dict['location'] = business['location']
            parsed_dict['category'] = business['categories']
            parsed_dict['id'] = business['id']
            parsed_dict['coordinates'] = business['coordinates']
            parsed_results.append(parsed_dict)
    except:
        pass

    return parsed_results
```
